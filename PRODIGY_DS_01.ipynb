```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("population.csv")
```

```python
aruba_data = df[df['Country Name'] == 'Aruba']

# Plotting a bar chart
plt.figure(figsize=(12, 6))
plt.bar(aruba_data['Year'], aruba_data['Value'], color='blue')
plt.title('Population Distribution in Aruba (1960-2021)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

```python
zimbabwe_data = df[df['Country Name'] == 'Zimbabwe']

# Plotting a histogram
plt.figure(figsize=(12, 6))
plt.hist(zimbabwe_data['Value'], bins=20, color='green', edgecolor='black')
plt.title('Population Distribution in Zimbabwe (1960-2021)')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```
